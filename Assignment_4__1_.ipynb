```python
import pandas as pd

patient_df=pd.read_csv(r"C:\Users\NIKHIL\Downloads\data science assignment\data science assignment\Patient_Data.csv")
billing_df=pd.read_csv(r"C:\Users\NIKHIL\Downloads\data science assignment\data science assignment\Billing_Data.csv")

patient_df.info()
billing_df.info()
patient_df.describe()

patient_df.columns

billing_column=['PatientID','Department','Doctor','BillAmount']
patient_bill=patient_df[billing_column]
print(patient_bill)

drop_col=['ReceptionistID','CheckInTime']
patient_df=patient_df.drop(columns=drop_col,errors="ignore")
print(patient_df)

dept_revenue=patient_bill.groupby("Department")["BillAmount"].sum()
print(dept_revenue)

patient_df=patient_df.drop_duplicates(subset="PatientID")
print(patient_df.shape)

mean_bill=patient_df['BillAmount'].mean()
print(mean_bill)

patient_df['BillAmount']=patient_df['BillAmount'].fillna(mean_bill)
print(patient_df)

merge_df=pd.merge(
    patient_df,
    billing_df,
    on="PatientID",
    how="inner")
print(merge_df)

new_patient=pd.DataFrame({
    "PatientID" : [106,107],
    "Name" :['Radhe','Sham'],
    "Department" :['Neurology','Orthopedics'],
    "Doctor" :['Dr. Smith','Dr. Manu'],
    "BillAmount" :[2000,3500]})
final_df=pd.concat([merge_df,new_patient],ignore_index=True)

final_df.loc[final_df["PatientID"]==106,"InsuranceCovered"]=500
final_df.loc[final_df["PatientID"]==107,"InsuranceCovered"]=1200

final_df["FinalAmount"]=(final_df["BillAmount"]-final_df["InsuranceCovered"])

print(final_df)



```
