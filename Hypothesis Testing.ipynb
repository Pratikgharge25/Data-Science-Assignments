```python
#Q(1)
import pandas as pd
import numpy as np
from scipy import stats

df= pd.Series([
    1.13, 1.55, 1.43, 0.92, 1.25, 
    1.36, 1.32, 0.85,1.07, 1.48, 
    1.20, 1.33,1.18, 1.22, 1.29
    ])

df.info()

n=len(df)
print("Sample size :",n)

mean=df.mean()
print("Sample Mean :",round(mean,4))

s=df.std()
print("Sample Std Deviation :",round(s,4))

alpha =0.01

dfree=n-1

t_crit=stats.t.ppf(1-alpha/2,dfree)
se_t=s/np.sqrt(n)
margin_t=t_crit*se_t
ci_t = (mean - margin_t, mean + margin_t)

sigma=0.2
z_crit=stats.norm.ppf(1-alpha/2)
se_z=sigma/np.sqrt(n)
margin_z=z_crit*se_z
ci_z=(mean-margin_z,mean+margin_z)

print("\n99% CI using T-Distribution :")
print(tuple(round(x,4)for x in ci_t))

print("\n99% CI using Sigma=0.2 :")
print(tuple(round(x,4)for x in ci_z))



#Q(2)
import numpy as np
from scipy import stats

x_bar = 3050        
n = 25              

X_mean = 600
cost_fixed = 1000
cost_per_unit = 5

mu = cost_fixed + cost_per_unit * X_mean


sigma = 5 * 25
alpha = 0.05

print("HYPOTHESES:")
print("H0: μ = 4000  (weekly cost follows the model)")
print("H1: μ > 4000  (weekly cost is higher than the model)")
print()

z = (x_bar - mu) / (sigma / np.sqrt(n))

print("TEST STATISTIC:")
print("Z value =", round(z, 4))
print()

z_crit = stats.norm.ppf(1 - alpha)
p_value = 1 - stats.norm.cdf(z)

print("CRITICAL VALUE (Right-tailed, α=0.05):", round(z_crit, 4))
print("P-value:", round(p_value, 6))
print()

print("DECISION:")

if z > z_crit:
    print("Reject H0")
else:
    print("Fail to Reject H0")

print()


print("CONCLUSION:")

if z > z_crit:
    print("There is strong evidence that weekly operating costs are higher than predicted by the model.")
else:
    print("There is not enough evidence to support the claim that weekly operating costs are higher than the model.")















```
