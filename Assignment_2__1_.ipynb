```python
#Q3
import numpy as np
import pandas as pd

data = {
    "Name": ["pratik","Ayush","Saurabh","Gudya","Harshad","Viraj","Shreyas","Omkar","Yash","Radhe"],
    "Age": [21,22,23,25,24,31,21,22,23,24],
    "City": ["Pune","Mumbai","Mumbai","Satara","Sangli","Kolhapur","Akola","Nagpur","Beed","Thane"],
    "Salary": [np.nan,2000,3000,np.nan,4000,2000,5000,np.nan,5000,2500]
}

# Create DataFrame
df = pd.DataFrame(data)

# Find mean salary
mean_salary = df["Salary"].mean()

# Fill missing values with mean
df["Salary"] = df["Salary"].fillna(mean_salary)

print("\n-----Data Frame-----")
print(df)

print("\n-----Info-----")
print(df.info())

print("\n-----Description-----")
print(df.describe())

print("\n-----Name and City Columns-----")
print(df[["Name", "City"]])

# Dropping column
df_drop = df.drop("City", axis=1)

print("\n-----After Dropping City Column-----")
print(df_drop)

# Removing duplicate rows
df_duplicate = df.drop_duplicates()

print("\n-----After Removing Duplicate Rows-----")
print(df_duplicate)
```
